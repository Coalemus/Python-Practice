# Using all functions for matplotlib we could have, we get this:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(-10,10,1000)
y = np.linspace(-10,10,1000)

plt.plot(x, np.sin(x), color='b', linestyle='-.', label='sin(x)')
plt.plot(y, np.cos(x), color='r', linestyle='--', label='cos(x)')
plt.hlines(y=0, xmin=-10, xmax=10, color='k', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Basic Plot')
plt.legend() # add handles arg.

plt.show
```

# Example of plotting two lines with style arguments

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color='m', label='sin(x)')
plt.plot(x, np.cos(x), color='y', linestyle='--', label='cos(x)')
plt.xlabel('sin')
plt.ylabel('cos')
plt.title('Double functions tinkered with style')
plt.legend()
plt.show
```
